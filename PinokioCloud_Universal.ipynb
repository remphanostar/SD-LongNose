```python
#@title 🛠️ Universal Setup: Platform Detection & Clean Repository
#@markdown **Automatically detects your platform and sets up PinokioCloud optimally**

import os
import sys
import subprocess
import time
import json
from pathlib import Path
import shutil

class PlatformDetector:
    """Universal platform detection and optimization"""
    
    def __init__(self):
        self.info = self._detect()
        self.paths = self._setup_paths()
        
    def _detect(self):
        """Detect current platform"""
        # Google Colab
        try:
            import google.colab
            return {
                'platform': 'google_colab',
                'name': 'Google Colab',
                'venv': True,
                'conda': True,
                'tips': ['Full GPU support', 'All features available']
            }
        except ImportError:
            pass
        
        # Lightning AI
        if ('LIGHTNING' in str(os.environ) or '/teamspace/' in os.getcwd()):
            return {
                'platform': 'lightning_ai',
                'name': 'Lightning AI',
                'venv': False,
                'conda': False,
                'tips': ['Simple apps work best', 'Use --user installs']
            }
        
        # Vast.AI
        elif '/workspace' in os.getcwd():
            return {
                'platform': 'vast_ai',
                'name': 'Vast.AI',
                'venv': True,
                'conda': True,
                'tips': ['Full support', 'Persistent storage']
            }
        
        # Local fallback
        else:
            return {
                'platform': 'local',
                'name': 'Local Machine',
                'venv': True,
                'conda': True,
                'tips': ['Full features available']
            }
    
    def _setup_paths(self):
        """Setup platform-specific paths"""
        platform = self.info['platform']
        
        if platform == 'google_colab':
            base = Path('/content')
        elif platform == 'lightning_ai':
            base = Path('/teamspace/studios/this_studio')
        elif platform == 'vast_ai':
            base = Path('/workspace')
        else:
            base = Path.cwd()
        
        return {
            'base': base,
            'repo': base / 'SD-LongNose',
            'scripts': base / 'SD-LongNose' / 'PinokioCloud_Colab',
            'apps': base / 'pinokio_apps'
        }

# Detect platform
detector = PlatformDetector()
info = detector.info
paths = detector.paths

print("🔍 PLATFORM DETECTION")
print("=" * 50)
print(f"🌩️ Platform: {info['platform'].upper()}")
print(f"☁️ Provider: {info['name']}")
print(f"🐍 Virtual envs: {'✅' if info['venv'] else '❌'}")
print(f"🐍 Conda: {'✅' if info['conda'] else '❌'}")
print(f"💡 Tips: {', '.join(info['tips'])}")

# Setup paths
REPO_DIR = paths['repo']
SCRIPTS_DIR = paths['scripts']
APPS_DIR = paths['apps']

print(f"\n📁 PATHS")
print(f"📍 Repository: {REPO_DIR}")
print(f"📍 Scripts: {SCRIPTS_DIR}")
print(f"📍 Apps: {APPS_DIR}")

# Clone repository
REPO_URL = "https://github.com/remphanostar/SD-LongNose.git"
print(f"\n📥 CLONING REPOSITORY")
print(f"🌐 URL: {REPO_URL}")

if REPO_DIR.exists():
    print("🔄 Updating...")
    try:
        cwd = os.getcwd()
        os.chdir(str(REPO_DIR))
        result = subprocess.run(['git', 'pull'], capture_output=True, timeout=30)
        os.chdir(cwd)
        print("✅ Updated" if result.returncode == 0 else "⚠️ Update failed")
        if result.returncode != 0:
            shutil.rmtree(REPO_DIR)
    except:
        if REPO_DIR.exists():
            shutil.rmtree(REPO_DIR)

if not REPO_DIR.exists():
    try:
        subprocess.run([sys.executable, "-m", "pip", "install", "-q", "gitpython"], check=True)
        import git
        git.Repo.clone_from(REPO_URL, REPO_DIR, depth=1)
        print("✅ Cloned")
    except Exception as e:
        print(f"❌ Failed: {e}")
        raise

# Verify files
required = ["streamlit_colab.py", "unified_engine.py", "cleaned_pinokio_apps.json"]
print(f"\n🔍 VERIFICATION")
for file in required:
    path = SCRIPTS_DIR / file
    if path.exists():
        size = path.stat().st_size
        print(f"✅ {file}: {size:,} bytes")
    else:
        print(f"❌ {file}: MISSING")
        raise FileNotFoundError(f"Missing: {file}")

# Dependencies
deps = ["streamlit>=1.28.0", "requests>=2.28.0", "nest_asyncio>=1.5.0"]
if info['platform'] != 'lightning_ai':
    deps.extend(["pyngrok>=7.0.0", "psutil>=5.9.0"])

print(f"\n📦 DEPENDENCIES")
for dep in deps:
    try:
        cmd = [sys.executable, "-m", "pip", "install", "-q"]
        if info['platform'] == 'lightning_ai':
            cmd.append("--user")
        cmd.append(dep)
        subprocess.run(cmd, check=True, timeout=60)
        print(f"✅ {dep.split('>=')[0]}")
    except:
        print(f"⚠️ {dep.split('>=')[0]}")

# Final setup
try:
    import nest_asyncio
    nest_asyncio.apply()
except:
    pass

sys.path.insert(0, str(SCRIPTS_DIR))
APPS_DIR.mkdir(parents=True, exist_ok=True)

# Load apps database
with open(SCRIPTS_DIR / "cleaned_pinokio_apps.json") as f:
    apps_data = json.load(f)

print(f"\n🎉 READY!")
print(f"🌩️ {info['name']}")
print(f"📊 {len(apps_data)} apps")
print("\n➡️ Next: Configure ngrok & launch")
```

```python
#@title 🚀 Launch Revolutionary PinokioCloud with ngrok
#@markdown **Configure ngrok token and launch revolutionary interface**

#@markdown ### 🔑 ngrok Token (Optional)
#@markdown Get free token: https://ngrok.com → Sign up → Copy authtoken
NGROK_TOKEN = ""  #@param {type:"string"}

#@markdown ### 🌍 Configuration
REGION = "us"  #@param ["us", "eu", "ap", "au", "sa", "jp", "in"]
PORT = 8501  #@param {type:"integer"}

import subprocess
import time
import socket

def detect_gpu():
    """Detect GPU"""
    try:
        result = subprocess.run(['nvidia-smi', '--query-gpu=name', '--format=csv,noheader'], 
                              capture_output=True, text=True, timeout=5)
        if result.returncode == 0 and result.stdout.strip():
            gpu_name = result.stdout.strip().split('\n')[0]
            print(f"🎮 GPU: {gpu_name}")
            os.environ['CUDA_VISIBLE_DEVICES'] = '0'
            return gpu_name
    except:
        pass
    
    print("⚠️ No GPU detected - CPU mode")
    return "CPU Only"

def find_port(start=8501):
    """Find available port"""
    for port in range(start, start + 50):
        try:
            with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
                s.bind(('localhost', port))
                return port
        except:
            continue
    return start

def setup_ngrok(port):
    """Setup ngrok tunnel"""
    if not NGROK_TOKEN or not NGROK_TOKEN.strip():
        print("⚠️ No ngrok token - localhost only")
        print("💡 Get free token at https://ngrok.com for public access")
        return None
    
    try:
        from pyngrok import ngrok, conf
        
        print(f"🔑 Configuring ngrok...")
        ngrok.set_auth_token(NGROK_TOKEN.strip())
        conf.get_default().region = REGION
        
        print(f"🌐 Creating tunnel...")
        tunnel = ngrok.connect(port)
        
        print(f"✅ Public URL: {tunnel.public_url}")
        return tunnel.public_url
        
    except Exception as e:
        print(f"❌ ngrok failed: {e}")
        if "authtoken" in str(e):
            print("🔑 Invalid token - get correct token from https://ngrok.com")
        return None

def launch_ui():
    """Launch Revolutionary UI"""
    ui_file = SCRIPTS_DIR / "streamlit_colab.py"
    
    if not ui_file.exists():
        print(f"❌ UI missing: {ui_file}")
        return None
    
    port = find_port(PORT)
    print(f"🔌 Port: {port}")
    
    cmd = [
        sys.executable, "-m", "streamlit", "run", "streamlit_colab.py",
        "--server.port", str(port),
        "--server.headless", "true",
        "--server.enableCORS", "false",
        "--browser.gatherUsageStats", "false"
    ]
    
    print("🚀 Starting Revolutionary UI...")
    process = subprocess.Popen(cmd, cwd=str(SCRIPTS_DIR))
    
    time.sleep(8)  # Wait for startup
    
    if process.poll() is None:
        print("✅ UI started!")
        return process, port
    else:
        print("❌ UI failed")
        return None, None

# Run detection
gpu = detect_gpu()

# Platform optimization
if info['platform'] == 'google_colab':
    os.environ['COLAB_GPU'] = '1'
elif info['platform'] == 'lightning_ai':
    os.environ['LIGHTNING_AI'] = '1'

print(f"\n🚀 LAUNCHING REVOLUTIONARY PINOKIOCLOUD")
print("=" * 60)

# Launch
process, port = launch_ui()

if process:
    public_url = setup_ngrok(port)
    
    print("\n🎉 REVOLUTIONARY PINOKIOCLOUD LIVE!")
    print("=" * 50)
    
    if public_url:
        print(f"🌐 Public: {public_url}")
    print(f"📱 Local: http://localhost:{port}")
    print(f"🌩️ Platform: {info['name']}")
    print(f"🎮 GPU: {gpu}")
    
    print("\n🌟 REVOLUTIONARY FEATURES:")
    print("⭐ Live GitHub stars integration")
    print("🎨 Beautiful cyberpunk interface")
    print("📊 Quality scoring system")
    print("🖥️ Terminal v2.0 with search")
    print("🔧 Platform optimization")
    
    print("\n⚠️ Keep cell running!")
    
    try:
        while process.poll() is None:
            time.sleep(30)
    except KeyboardInterrupt:
        print("\n🛑 Stopping...")
        process.terminate()
        if public_url:
            try:
                from pyngrok import ngrok
                ngrok.disconnect(public_url)
            except:
                pass
        print("✅ Stopped")
else:
    print("❌ Launch failed")
    print(f"🔧 Troubleshooting for {info['platform']}:")
    
    if info['platform'] == 'lightning_ai':
        print("⚡ Try restarting studio")
        print("💡 Consider Google Colab for full compatibility")
    elif info['platform'] == 'google_colab':
        print("☁️ Runtime → Restart runtime")
        print("💡 Ensure GPU runtime enabled")
    else:
        print("🔧 Check dependencies and internet connection")
```
